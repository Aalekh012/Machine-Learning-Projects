### Train Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
x=df.drop("prognosis",axis=1)
```

```python
x
```

```python
y=df['prognosis']
```

```python
y
```

### Label Encoding the string values

```python
le=LabelEncoder()
le.fit(y)
Y=le.transform(y)
```

```python
Y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.3,random_state=20)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### Training some classification models to check which model give best results

```python
#For generateing synthetic datasets: importing make_classification:
from sklearn.datasets import make_classification

#Importing Support vector machine as one of our model to test its accuracy for the project:
from sklearn.svm import SVC

#Importing RandomForestClassifier and GradientBoostingClassifier to test their accuracy for the project:
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier

#Importing Multi Nomial Navie Bayes to test its accuracy for the project:
from sklearn.naive_bayes import MultinomialNB

#Also importing accuracy score and confusion matrxi to see the performance of different models we use:
from sklearn.metrics import accuracy_score,confusion_matrix
```

Creating a Dictionary to store all the models:

```python
models={
    "SVC":SVC(kernel='linear'),
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42),
    "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "MultiNomialNB":MultinomialNB()
}

for model_name,model in models.items():
    
    #training the model:
    model.fit(x_train,y_train)
    
    #testing the model:
    predictions=model.predict(x_test)
    
    #Calculating the accuracy:
    accuracy=accuracy_score(y_test,predictions)
    
    #Calculating the confusion matrix:
    cm=confusion_matrix(y_test,predictions)
    
    print(f"{model_name} Accuracy: {accuracy}")
    print("f{model_name} Confusion Matrix")
    print(np.array2string(cm,separator=', '))
```
